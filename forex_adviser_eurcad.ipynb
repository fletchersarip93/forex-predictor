## EURCAD Data

```python
import pandas as pd

df = pd.read_csv('data/EURCAD.csv', dtype={'<DTYYYYMMDD>': str,
                                          '<TIME>': str})
df = df.rename(columns={'<DTYYYYMMDD>': 'date',
                       '<TICKER>': 'ticker',
                       '<TIME>': 'time',
                       '<OPEN>': 'open',
                       '<HIGH>': 'high',
                       '<LOW>': 'low',
                       '<CLOSE>': 'close',
                       '<VOL>': 'vol'})
df['date_time'] = df['date'].str.cat(df['time'], sep='@')
```
